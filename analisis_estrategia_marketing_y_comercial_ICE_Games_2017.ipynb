### 2.3.1 Comentarios de la acción previamente realizada:

Al momento de verificar los datos en cuestión se llego a la conclusión de cambiar los datos de las columnas "year_of_release" a tipo de fecha y se analiza la opción de modificar la columna "user_score" a float, sin embargo al tener la abeviatura "TBD" se procedera a darle un trato especial para poder conservar los datos y modificar la columna a float al mismo tiempo. El como se explicara en el paso siguiente.

## 2.4 Trato a los datos ausentes

```python
# Se cambian los datos ausentes de la columna "genre" por la palabra "tbd"

df_games['genre'].fillna('tbd', inplace=True)

# Se contabiliza la mayor cantidad de datos que se repite de un valor y se ordena de mayor a menor incluyendo los datos ausentes y "tbd" los datos unicos que se encuentran "user_score" para analizar el metodo apropiado para tratar los datos ausentes y los "tbd"

df_games['user_score'].value_counts(dropna=False).sort_values(ascending=False)

# Se creara una nueva columna replicando la de "user_score", la nueva columna se llamara "user_score_replace_null" y en esta se cambiaran todos los datos "tbd" por datos ausentes, asi se conservara la columna original para mantener la relación de estos datos y a su vez se usara los "NaN" para poder convertir la columna nueva a float y no afectar los calculos necesarios para mas adelante:

df_games['user_score_replace_null'] = df_games['user_score']
df_games.loc[df_games['user_score_replace_null'] == 'tbd', 'user_score_replace_null'] = np.nan
df_games['user_score_replace_null'] = df_games['user_score_replace_null'].astype(float)

# Se cambian los los datos ausentes de la columna "raiting" por la palabra "tbd"

df_games['rating'].fillna('tbd', inplace=True)

# Se imprime los datos para verificar los cambios:

df_games.info()
print(' ')
print('----------------------------------------------------------------------')
print(' ')
df_games.head()
```

### 2.4.1 Comentarios de la acción previamente realizada:

El DF original cuenta con 10 columnas, de estas en 5 hay datos ausentes, y son las siguientes:

* year_of_release: esta columna es del tipo de fecha.
* genre: esta columna corresponde a los generos de los juegos y es string.
* critic_score: esta corresponde a la calificacion que le dio la critica a los juegos y es float.
* user_score: esta corresponde a la calificacion que le dio los usuarios a los juegos y es float.
* rating: esta es la clasificación de los juegos y es string.

El dato "tbd" (to be determinated) hace referencia a que "el dato se encuentra en proces para determinar tal o cual acción, correspondiente al contexto el cual referencie", osease, si se usa en "genre" hace referencia a que "se esta determinando que genero es el juego", en cuanto a la critica "que calificación se le dara" y así sucesivamente, se explica ya que en los parrafos siguientes sera relevante esta información.

* De las columnas anteriores la columna "year_of_release" no se le modificaron los datos ausentes, ya que, al hacerlo en automatico se cambian el tipo de datos de "date" a "string" y esto podria alterar los estadisticos mas adelante.

* De la columna "genre" se procedio a modificar los valores ausentes con la abreviatura "tbd" ya que, en esta columna al no estar determinado o estar marcado un juego como dato ausente, es de predisponerse a que "se esta determinando el dato" por tal motivo se usa esa abreviatura en ingles para poder tener la referencia de este dato.

* En la columna "critic_score" se registro desde el principio como "float" por tal motivo todos los campos se registraron como datos ausentes, la relevancia de este datos corresponde a que "se puede usar esta columna como float aun cuando se cuente con datos ausentes, sin embargo si estos se cambian por un string en automatico se transforma a float y deja de ser una columna util para operaciones aritmeticas", para ello se procedio a analizar en primer instancia "cuantos datos ausentes tiene esta columna" y resulto que "un aproximado del 50% de los datos se encuentran ausentes", y en segunda instancia "los datos se encuentras un tanto dispersos", por tal motivo, sim importar que los datos ausentes de esta columna sean llenados por la media o la mediana de la columna "los datos seran sesgados y no presentaran un estadistico claro al momento de realizar el analisis", por tal motivo, para salvaguardar de manera correcta la integridad de la información y poder tener una muestra clara y real de la distribución de los datos, se procedera a dejar los datos ausentes de esta manera, las operaciones necesarias se realizaran con estos datos ausentes para no sesgar el estaistico correspondiente mas adelante.

* Relacionado a la columna "user_score", se descubrio que esta columna, a pesar de ser en su mayoria un float, varios de sus valores estaban registrados con la abreviatura "tbd", por tal motivo esto genero que esta columna sea del tipo "string", así que, para conservar el dato exacto de "cuantos juegos ya tenian calificacion pero estaban pendientes de colocar, los datos ausentes y los no ausentes", se procedio a crear una nueva columna con el nombre "user_score_replace_null" en la cual se cambiaron todos los "tbd" por datos ausentes para así poder convertir la columna el float y poder realizar operaciones con ella, ahora, cuando se obtuvo la conversión de los datos ausentes, se edecubrio que "mas del 55% de los datos totales son ausentes, y aunque estos datos se encuentren un poco mas centrados, al haber mas del 50% de datos ausentes, sin importar si se cambian los datos ausentes con la media o mediana, los datos seran sesgados y no representara de manera real la calificación de los usuarios, por tal motivo, al igual que en el caso de la columna "critic_score" se procedera a dejar los datos ausentes de esa manera en la nueva columna (user_score_replace_null).

* Por ultimo, en la columna "raiting" se procedio a acambiar los datos ausentes por la abreviatura "tbd" siguiendo el mismo caso de la columna "genre".

## 3.4 ¿Cuanto tiempo tardan en aparecer y desaparecer las plataformas?

[En el siguiente apartado se plazmara la vida de las plataformas]

```python
# Se genera un nuevo df con el nombre "platform_lifespan" tomando las columnas "platform" y "year_of_release_int" del df "df_games" para saber el tiempo de vida de las plataformas sin modificar el df "df_games"

platform_lifespan = df_games.groupby('platform')['year_of_release'].agg(['min', 'max']).reset_index()

# Se crea la columna columna "lifespan" con la cantidad de años que estuvieron vivas las plataformas en el mercado:

platform_lifespan['lifespan'] = platform_lifespan['max'] - platform_lifespan['min']

# Se ordena las plataformas:

platform_lifespan = platform_lifespan.sort_values(by='lifespan', ascending=False)

# Se imprime información del DF para verificación de datos:

platform_lifespan.info()
print(' ')
print('----------------------------------------------------------------------')
print(' ')

# Se genera un grafico de gantt con el df "platform_lifespan"
# Primero se generan las columnas para la generación apropiada de la grafica:

platform_lifespan['start'] = platform_lifespan['min']
platform_lifespan['end'] = platform_lifespan['max']

# Se genera la grafica:

fig = px.timeline(platform_lifespan, x_start="start", x_end="end", y="platform", color="platform")
fig.update_yaxes(categoryorder="total ascending")
fig.update_layout(title="Duración de vida de plataformas (Gráfico tipo Gantt)")
fig.show()
```

# 5 Pruebas de Hipotesis

## 5.1 Calificaciones promedio Xbox One VS PC

[En este punto se verificara la hipotesis de si las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.]

```python
# Prueba de hipotesis:

# H0 = Las calificaciones promedio de los usurios para las pltaformas "xone" y "pc" son iguales.
# H1 = Las califiaciones promedio de los usuarios para las plataformas "xone" y "pc" son diferentes.

# Se obtiene el valor de interes, el cual se almacenara en 2 variables, una diseñada para almacenar el total de las calificaciones de los usuarios sobre la plataforma "xone" y la otra para almacenar el total de las calificaciones de los usuarios sobre la plataforma "pc" del dataframe "df_games"

xone_ratings = df_games[df_games['platform'] == 'xone']['user_score_replace_null']
pc_ratings = df_games[df_games['platform'] == 'pc']['user_score_replace_null']

# Se cambian los datos nulos de xone_ratings por su mediana y los de pc_ratings por la media:

xone_ratings = xone_ratings.fillna(xone_ratings.median())
pc_ratings = pc_ratings.fillna(pc_ratings.mean())

# Se visualiza el promedio de las calificaciones de ambas plataformas:

print("El valor de interes es el siguiente:")
print(f"Xone: {xone_ratings.mean()}")
print(f"PC: {pc_ratings.mean()}")

# Se determina el valor del umbral:

alpha = 0.05
print(' ')
print('----------------------------------------------------------------------')
print(' ')
print(f"El valor del umbral es: {alpha}")

# Se realiza la prueba t de student para muestras independientes:

results = st.ttest_ind(xone_ratings, pc_ratings)

# Se imprime el valor "p-value":

print(' ')
print('----------------------------------------------------------------------')
print(' ')
print(f"El valor del p-value es: {results.pvalue}")

# Se compara el "p-value" con el umbral:

if (results.pvalue < alpha):
    print("Se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.")
    print(' ')
    print('----------------------------------------------------------------------')
    print(' ')
else:
    print("No se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son iguales.")
```

###5.1.1 Comentarios sobre el punto 5.1

R: Para probar la hipotesis nula se utilizo una prueba t de student para muestras independientes, para así poder comparar las medias de ambas variables (media de la calificación de los usuarios de Xbox One y PC). Cabe resaltar que, al momento de realizar el calculo de la media de ambas plataformas, primero se tuvieron que sustituir los datos nulos, en la consola Xbox One resultaron 64 datos nulos aproximadamente y en PC resultaron mas de 200, así que, para poder realizar la prueba t, primero se procedio a analizar la distribución de los registros en ambas variables, y se descubrio que en la plataforma de Xbox One estaba mas uniforme la distribución entre sus multiples registros, por tal motivo se opto por manejar la mediana, en cambio los resultados de PC mostraron que la mayoria de los usuarios se concentraban en los valores +6.5 - +8.5 por tal motivo se procedio a usar la media para sustituir los datos nulos.

Antes de hacer el cambio anteriormente mencionado, se procedio a verificar la media de ambos registros, cuando se hizo esto los valores resultantes previos a sustituir los datos nulos dieron una cifra similar variando unicamente por milesimas el dato final, por tal motivo se determino que el manejo de los datos nulos de ambas variables se llevo a cabo de la manera mas adecuada.

## 5.2 Calificaciones promedio generos "Acción" VS "Deportes".

[En este punto se verificara la hipotesis de Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.]

```python
# Prueba de hipotesis:

# H0 = Las calificaciones promedio de los usuarios para los generos de "action" y "sports" son diferentes.
# H1 = Las calificaciones promedio de los usuarios para los generos de "action" y "sports" son iguales.

# Se obtiene el valor de interes, el cual se almacenara en 2 variables, una diseñada para almacenar el total de las calificaciones de los usuarios sobre el genero "action" y la otra para almacenar el total de las calificaciones de los usuarios sobre el genero "sports" del dataframe "df_games"

action_ratings = df_games[df_games['genre'] == 'action']['user_score_replace_null']
sports_ratings = df_games[df_games['genre'] == 'sports']['user_score_replace_null']

# Se calculan los datos nulos de cada una de las variables:

action_null = df_games[df_games['genre'] == 'action']['user_score_replace_null'].isna().sum()
sports_null = df_games[df_games['genre'] == 'sports']['user_score_replace_null'].isna().sum()

# Se imprime los datos nulos:

print('La cantidad de datos nulos que hay en cada variable previo a determinar por que valor se cambiaran estos mismos es el siguiente:')
print("Los datos nulos de action son: ", action_null)
print("Los datos nulos de sports son: ", sports_null)

# Se imprime el promedio de las variables "action_ratings" y "sports_ratings" antes de cambiar los datos nulos:
print(' ')
print('----------------------------------------------------------------------')
print(' ')
print("El promedio de las variables respectivamente es el siguiente:")
print(f"Action: {action_ratings.mean()}")
print(f"Sports: {sports_ratings.mean()}")

# Se opta por cambiar los datos nulos por el promedio de cada variable respectivamente:

action_ratings = action_ratings.fillna(action_ratings.mean())
sports_ratings = sports_ratings.fillna(sports_ratings.mean())

# Se imprime el promedio de las variables "action_ratings" y "sports_ratings" despues de cambiar los datos nulos

print(' ')
print('----------------------------------------------------------------------')
print(' ')
print("El valor de interes es el siguiente:")
print(f"Action: {action_ratings.mean()}")
print(f"Sports: {sports_ratings.mean()}")

# Se determina el valor del umbral:

alpha = 0.05
print(' ')
print('----------------------------------------------------------------------')
print(' ')
print(f"El valor del umbral es: {alpha}")

# Se realiza la prueba t de student para muestras independientes:

results = st.ttest_ind(action_ratings, sports_ratings)

# Se imprime el valor "p-value":

print(' ')
print('----------------------------------------------------------------------')
print(' ')
print(f"El valor del p-value es: {results.pvalue}")

# Se compara el valor "p" con el umbral:

if (results.pvalue < alpha):
    print("Se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.")
    print(' ')
    print('----------------------------------------------------------------------')
    print(' ')
else:
    print("No se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son iguales.")
```

### 5.2.1 Comentarios sobre el punto 5.2

R: Con base a los requerimientos del proyecto, se elaboraron las hipotesis tanto nula como alternativa. Al reevisar los registros se pudo apreciar una cantidad superior a los mil registros de datos nulos, así que se procedio a elaborar tanto un diagrama de caja como otro tipo de graficos previos a la elaboración de la linea de codigo de la hipotesis, al elaborar estos graficos se pudo apreciar que la alternativa mas viable para no dañar la distribucion de los datos era "sustituir los datos nulos por el promedio. De esta manera se procedio a desarrollar este paso.

*** Estrategia comercial: ***

1. Se recomienda segmentar el catalogo de juegos por regiones, para ahorrar la distribución del mercado, se dividiria el filtro en 2, uno para occidente y otro para oriente.
2. Para el catalogo de juegos de occidente, se recomienda hacer convenios directos con los distintos estudios para poder pactar contratos de exlusividad, enfonandonos en estudios que esten diseñando juegos tipo shooters, acción y de deportes, especialmente para las consolas de Play Station 4 y Xbox One, ya que esta combinación fue la que presento mayor volumen de facturación durante el año 2016.
3. los juegos con mayor tiket promedio fueron aquellos con clasificación para adultos, por lo tanto, se recomienda anexar juegos con la modalidad "free to pay", enfocados para adultos (retomando la estrategia de juegos tipo "shooters, accion y deportes"), para las consolas de Play Statión 4 y Xbox One, donde el procesamiento de pago pase por nuestros servidores, para así poder recibir utilidades por cada transacción realizada.
4. En el caso del mercado oriental, el caralogo de juegos estaria enfocado mas en juegos de los generos de "accion, roles, misc, deportes y disparos", dandole prioridad a los primeros 3, al igual que en occidente, se recomienda focalizar esfuerzos para firmar contratos de exclusividad con los estudios respectivos encargados de la creación de juegos de este tipo, para así poder tener mayores ingresos que nuestra competencia.
5. En el caso de Asía, los juegos con mayor volumen de facturación fueron aquellos juegos familiares y categorizados como "de todas las edades", por tal motivo se recomienda buscar juegos del tipo familiar para el catalogo de Play Station y Nintendo 3D´s.
